```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# reading the csv file

df = pd.read_csv('jumta.csv')

df.head()
```

```python
df.info()
```

```python
# sample code 
x = np.linspace(0,10,100)
y = np.sin(x)

plt.plot(x,y)

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('A simple line plot')

plt.show()
```

```python
# Create a line plot showing the monthly revenue over the two-year period. Customize the plot to make it publication-quality.


x = df['Month']
y = df['Revenue']

plt.figure(figsize=(10,5))

plt.plot(x,y, marker='o',linestyle='-',linewidth=2,label='Revenue')

plt.xlabel('Month')
plt.ylabel('Revenue')
plt.grid()
plt.legend()
plt.xticks(rotation = 45)
plt.show()
```

```python
# Create a bar plot showing the total number of orders per month. Customize the plot to make it publication-quality.


plt.figure(figsize=(10,9))

plt.bar(df['Month'], df['Orders'])
plt.title('Orders of The Month')
plt.xlabel('Month')
plt.ylabel('Oders')
plt.xticks(rotation=45)

plt.show()
```

```python
# Create a scatter plot showing the relationship between the number of orders and the average order value. Customize the plot to make it publication-quality.

plt.figure(figsize=(10,9))
plt.scatter(df['Orders'], df['Average_Order_Value'])

plt.show()
```
